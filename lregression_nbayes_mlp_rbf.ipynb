```python
# imports
import pandas as pd
import numpy as np
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import explained_variance_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn import metrics
from sklearn.model_selection import cross_val_predict
from sklearn import model_selection
from scipy.stats import ttest_ind
from statsmodels.stats.multicomp import MultiComparison
from statsmodels.stats.multicomp import pairwise_tukeyhsd
from scipy import stats
#
from sklearn.linear_model import LogisticRegression
#
from sklearn.naive_bayes import GaussianNB
#
from sklearn.neural_network import MLPClassifier
#
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
```

```python
# Base 1 - Career Longevity for NBA Rookies (preve a chance de um atleta ter uma carreira com mais de 5 anos e menos de 5 anos)

df1 = pd.read_csv('nba_logreg.csv').dropna() # data frame Career Longevity for NBA Rookies

x_1 = df1[['GP', 'MIN', 'PTS','FGM']]

y_1 = df1['TARGET_5Yrs']

x_1.describe()
```

```python
# Base 2 - Breast Cancer (detecção de cancer)

df2 = pd.read_csv('dataR2.csv').dropna()# data frame Breast Cancer

x_2 = df2[['Age','Glucose','Insulin','Adiponectin','Resistin']]

y_2 = df2['Classification']

x_2.describe()
```

```python
# Base 3 - Titanic Surviver (Predição da chance de sobrevivência no Titanic)

df3 = pd.read_csv('train.csv').dropna() # data frame Titanic Surviver

x_3 = df3[['Pclass', 'Sex', 'Age']]

x_3['Sex'].replace(to_replace=['female'], value=0, inplace=True)
x_3['Sex'].replace(to_replace=['male'], value=1, inplace=True)

y_3 = df3['Survived']

x_3.describe()
```

```python
# Naive Bayes

gnb = GaussianNB()

#
modelo_gnb_1 = gnb.fit(x_1,y_1)
modelo_gnb_2 = gnb.fit(x_2,y_2)
modelo_gnb_3 = gnb.fit(x_3,y_3)
```

```python
# Validação

def validacao(X,y, modelo):
    X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.4,random_state=0)


    # Train the model using the training sets
    modelo.fit(X_train,y_train)

    #Predict Output
    y_pred = modelo.predict(X_test)

    #matriz de confusão e metricas
    cnf_matrix = confusion_matrix(y_test, y_pred)

    specificity = cnf_matrix[1,1]/(cnf_matrix[1,0]+cnf_matrix[1,1])

    print("\nMétodo validação 60/40 resutados:")
    print("\nMatriz de confusão:\n",cnf_matrix,'\n')
    print("Accuracy:",accuracy_score(y_test, y_pred))
    print("Precision:",precision_score(y_test, y_pred))
    print("Specificity:", specificity)
    print("Recall:",metrics.recall_score(y_test, y_pred))
    print("F1_score:",metrics.f1_score(y_test, y_pred))
```

```python
# Validação Base 1

lreg_nbayes_mlp_rbf_testes(x_1,y_1,"Base 1",validacao)
```

```python
def t_test_validation(X,y, modelo, initial=1, final=11):
    t_test_validation_vector = []
    for x in range(initial,final):
        X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=x)
        modelo.fit(X_train, y_train)
        predito = modelo.predict(X_test)
        t_test_validation_vector.append(accuracy_score(y_test, predito))
        
    return t_test_validation_vector
```

```python
def t_test_validation_anova_tukey(x,y,base):   
    t_test_validation_reg = t_test_validation(x,y,reg)
    t_test_validation_gnb = t_test_validation(x,y,gnb)
    t_test_validation_mlp = t_test_validation(x,y,mlp)
    t_test_validation_mlp_lbfgs = t_test_validation(x,y,mlp_lbfgs)
    t_test_validation_rbf = t_test_validation(x,y,rbf)
    t_test_validation_rbf_gpc = t_test_validation(x,y,rbf_gpc)
    
    fvalue, pvalue = stats.f_oneway(t_test_validation_reg,
                                    t_test_validation_gnb, 
                                    t_test_validation_mlp,
                                    t_test_validation_mlp_lbfgs,
                                    t_test_validation_rbf,
                                    t_test_validation_rbf_gpc)
    
    print("---------------------------------------------------------------------")
    print("Teste ANOVA ("+base+")")
    print("---------------------------------------------------------------------")
    print()
    print("F-statistic:",fvalue)
    print("p-value:",pvalue)
    print()
    
    v = np.concatenate([t_test_validation_reg,t_test_validation_gnb,t_test_validation_mlp,t_test_validation_mlp_lbfgs,t_test_validation_rbf,t_test_validation_rbf_gpc])
    labels = ['reg']*len(t_test_validation_reg)+['gnb']*len(t_test_validation_gnb)+ \
     ['mlp']*len(t_test_validation_mlp)+['mlp_lbfgs']*len(t_test_validation_mlp_lbfgs)+ \
     ['rbf']*len(t_test_validation_rbf)+['rbf_gpc']*len(t_test_validation_rbf_gpc)
    
    tukey_results = pairwise_tukeyhsd(v, labels, 0.05)
    
    print("---------------------------------------------------------------------")
    print("Teste Tukey ("+base+")")
    print("---------------------------------------------------------------------")
    print(tukey_results)
```

```python
# Teste ANOVA e Tukey - Base 1

t_test_validation_anova_tukey(x_1,y_1, "Base 1")
```

```python
# Teste ANOVA e Tukey - Base 2

t_test_validation_anova_tukey(x_2,y_2, "Base 2")
```

```python
# Teste ANOVA e Tukey - Base 3

t_test_validation_anova_tukey(x_3,y_3, "Base 3")
```
